```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df=pd.read_csv("IEA-EV-dataEV salesHistoricalCars.csv")
```

```python
print(df.isnull().sum())
print(df.duplicated().sum())
```

```python
plt.boxplot(df['value'])
plt.show()
Q1=df['value'].quantile(0.25)
Q3=df['value'].quantile(0.75)
IQR=Q3-Q1
Lower_bound=Q1-1.5*IQR
Upper_bound=Q3+1.5*IQR
df=df[(df['value']>=Lower_bound)&(df['value']<=Upper_bound)]

plt.boxplot(df['value'])
plt.show()
```

```python
print(df.info())
print(df['value'].describe())
```

```python
df_country_classification=df.groupby(['region'])['region'].value_counts().sort_values(ascending=False)
#print(df_country_classification)
plt.bar(x=df_country_classification.index,height=df_country_classification.values,color='magenta')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
df_parameter_classification=df.groupby(['parameter'])['parameter'].value_counts().sort_values(ascending=False)
print(df_parameter_classification)
plt.bar(x=df_parameter_classification.index,height=df_parameter_classification.values)
plt.xticks(rotation=90)
plt.show()
```

```python
df_parameter_classification=df.groupby(['parameter'])['parameter'].value_counts().sort_values(ascending=False)
print(df_parameter_classification)
sns.countplot(df['parameter'])
plt.xticks(rotation=90)
plt.show()
```

```python
df_yearwise_sales=df.groupby(['year'])['value'].sum().sort_values(ascending=False)
print(df_yearwise_sales)
plt.bar(x=df_yearwise_sales.index,height=df_yearwise_sales.values)
plt.xticks(rotation=90)
plt.show()
```
